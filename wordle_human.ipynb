# Un juego de palabras

Es un juego en el se trata de adivinar palabras de una longitud determinada con las pistas que se van propocionando.

Lo primero que vamos a hacer es una función que se llame <code>read_words</code> que reciba como parámetros:
* el nombre del fichero 
* la longitud de palabra deseada 

La función devuelve una lista con las palabras del fichero de la longitud deseada.

```python
def read_words(filename,len_words):
    lst=[]
    with open(filename, 'r', encoding='utf-8', errors='ignore') as file:
        line=file.readline()
        while line!='':
            separate_line=line.strip()
            if len(separate_line)==len_words:
                lst.append(separate_line)
            line=file.readline()
    return lst
#Genero una lista con todas las palabras del fichero que tengan la longitud pedida, añadiendo solo las que tengan la longitud correcta
```
